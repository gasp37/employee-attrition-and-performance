Pudemos verificar que algumas colunas possuem valores qualitiativos que precisam ser transformatos em valores numericos

```python
df['BusinessTravel'].value_counts()
```

```python
col_bin = ['Attrition', 'Over18', 'OverTime', 'Gender']
df_bin = df.copy()


for v in col_bin:
    df_bin[v].replace({'Yes':1, 'Y':1, 'No':0, 'N':0, 'Female':1, 'Male':2}, inplace=True)
```

```python
# Importamos o OrdinalEncoder para transformar os valores das colunas qualitativas em valores numericos

from sklearn.preprocessing import OrdinalEncoder

col_cat = ['BusinessTravel', 'Department','EducationField','JobRole','MaritalStatus']

enc = OrdinalEncoder()
labled = enc.fit_transform(df_bin[col_cat])

df_bin.loc[:,col_cat] = labled
```

```python
enc.categories_
```

```python
df_bin.iloc[:7,:20]
```

```python
cols_with_std_zero = df_bin.columns[df_bin.std(axis=0) == 0]
df_no_constants = df_bin.drop(columns = cols_with_std_zero)
```

```python
corr = df_no_constants.corr()
corr = (corr)

f, ax = plt.subplots(figsize=(15,10))

sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            cmap='Spectral')
```

#### Observações sobre a correlação das variaveis:

Variaveis que possuem uma **correlação forte** com a **saida de um funcionario**:
- Trabalhar horas extras
- Pouco tempo no mercado de trabalho
- Cargo de baixo nivel
- Receita mensal baixa
- Idade
- Envolvimento no trabalho
- Satisfação com o trabalho
    

A **Idade** possui uma forte influência no turnover dos funcionários:
- Pessoas mais velhas e com **mais tempo no mercado de trabalho** possuem **salarios maiores**, ocupam **cargos mais altos** e com isso tem **menor chance de deixar a empresa**

--- 
Curiosidade:
Provavelmente, um típico caso de correlação != causalidade:
- A correlação entre StockOptionLevel x MaritalStatus tem um forte destaque no mapa, possivelmente, uma coincidência em que todos os colaboradores que possuem ações da companhia são casados e divorciados, e todos os solteiros não possuem ações da companhia.
